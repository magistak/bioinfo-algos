```python
def HammingDistance(st1:str, st2:str):
    if len(st1) != len(st2):
        raise Exception("Strings must be of equal length.")
    dist=[a for a,b in zip(st1,st2) if a!=b]
    return len(dist)
```

```python
def approx_match(kmer,Str,d):
    i=0
    pos=[]
    for _ in range(len(Str)):
        try: 
            ss=Str[i:i+len(kmer)]
            hd=HammingDistance(kmer,ss)
            # print(f"i: {i}")
            # print(f"hd: {hd}")
            # print(hd<=d)
            if hd<=d:
                pos.append(i)
            i+=1
        except: break
    return pos
sol=approx_match(kmer,Str,d)
" ".join(map(str,sol))
```

```python
def count_d(kmer,Str,d):
    return len(approx_match(kmer,Str,d))
```
